```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
data = pd.read_csv('D:\\Collage\\Graduation Project\\my_Data.csv')
```

```python
data
```

```python
X = data[['Age', 'Height', 'Weight', 'Gender', 'Body_Part']]
y = data['Exercise Name']
```

```python
X_encoded = pd.get_dummies(X, columns=['Body_Part'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
```

```python
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier.fit(X_train, y_train)
```

```python
y_pred = rf_classifier.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')
```

```python
print(classification_report(y_test, y_pred))
```

```python
# Input from the user
Age = float(input("Enter age: "))
Height = float(input("Enter height in cm: "))
Weight = float(input("Enter weight in kg: "))
Gender =float(input("Enter gender (male=0/female=1): "))
Body_Part = input("Enter body part (shoulders/neck/quadriceps/lower back): ").lower()
```

```python
user_data = pd.DataFrame({
    'Age': [Age],
    'Height': [Height],
    'Weight': [Weight],
    'Gender': [Gender],
    'Body_Part': [Body_Part]
    })
```

```python
user_data_encoded = pd.get_dummies(user_data, columns=['Body_Part'])
```

```python
missing_cols = set(rf_classifier.feature_names_in_) - set(user_data_encoded.columns)
for col in missing_cols:
    user_data_encoded[col] = 0
```

```python
user_data_encoded = user_data_encoded[rf_classifier.feature_names_in_]
```

```python
exercise_prediction = rf_classifier.predict(user_data_encoded)
```
